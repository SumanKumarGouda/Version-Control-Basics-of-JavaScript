```python
#Question 3: What is the purpose of a .gitignore file in a Git repository? Provide an example of how to ignore all .log files.
Answer-The .gitignore file tells Git which files or folders should NOT be tracked in a repository.
These are usually:
1.Temporary files
2.Log files
3.Build outputs
4.Cache files
5.Environment / secret files

In short: it keeps your repository clean and secure.

.gitignore Is Important:-
1.Prevents Unnecessary Files from Being Tracked.
2.Improves Security Sensitive files should never be pushed to GitHub.
3.Reduces Repository Size.


Example: Ignore All .log Files:-
Create or edit a .gitignore file and add: *.log
```
